```python
# Question 1: Descriptive analysis of your data

# STEP 1 - Load the dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)

# Load dataset
df = pd.read_csv("UCI_Credit_Card.csv")

# Basic info
print("Shape of dataset:", df.shape)
print("\nDataset information:")
df.info()

# Display first rows
df.head()

# STEP 2 - Descriptive statistics

# Basic statistics for numerical columns
desc = df.describe().T
display(desc)

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Check unique values for categorical variables
cat_vars = ["SEX", "EDUCATION", "MARRIAGE"]
for col in cat_vars:
    print(f"\n{col} value counts:")
    print(df[col].value_counts())


# STEP 3 - Distributions and histograms

num_features = ["LIMIT_BAL", "AGE"] + [f"PAY_{i}" for i in range(0,7) if f"PAY_{i}" in df.columns]

# Histograms
df[num_features].hist(figsize=(15,10), bins=30, edgecolor='black')
plt.suptitle("Distributions of numerical features", fontsize=16)
plt.show()

# Distribution of the target
sns.countplot(x="default.payment.next.month", data=df)
plt.title("Target variable distribution (default payment next month)")
plt.xlabel("Default (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.show()


# STEP 4 - Correlation analysis

# Compute correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Display correlation with the target variable
target_corr = corr_matrix["default.payment.next.month"].sort_values(ascending=False)
print("Correlation with target variable:\n")
print(target_corr.head(10))

# Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap")
plt.show()


# STEP 5 - Summary of findings

print("""
- The dataset contains 30,000 records and 25 variables.
- No missing values were detected.
- 22.1% of clients defaulted on their payment next month (imbalanced target).
- Variables such as PAY_0 to PAY_6, LIMIT_BAL, and BILL_AMT1-6 show strong variance.
- Some categorical features (EDUCATION, MARRIAGE) contain out-of-range values (e.g., 0, 5, 6).
- PAY_0 (most recent payment delay) shows the strongest correlation with default risk.
""")


```

```python
# Question 2: Implementation of the necessary pre-processing

# STEP 1 - Cleaning categorical features

# Check unique values before cleaning
print("Unique values before cleaning:")
for col in ["EDUCATION", "MARRIAGE"]:
    print(f"{col}: {sorted(df[col].unique())}")

# Recode incorrect values
df["EDUCATION"] = df["EDUCATION"].replace({0: 4, 5: 4, 6: 4})  # Group unknowns (0,5,6) into 4 = 'Other'
df["MARRIAGE"] = df["MARRIAGE"].replace({0: 3})                # Replace 0 by 3 = 'Other'

# Check after cleaning
print("\nUnique values after cleaning:")
for col in ["EDUCATION", "MARRIAGE"]:
    print(f"{col}: {sorted(df[col].unique())}")


# STEP 2 - Separate features and target

# Target variable
target = "default.payment.next.month"

# Drop identifier
X = df.drop(columns=["ID", target])
y = df[target]

print("X shape:", X.shape)
print("y shape:", y.shape)

# STEP 3 - Feature scaling (Standardization)

from sklearn.preprocessing import StandardScaler

# Instantiate scaler
scaler = StandardScaler()

# Fit and transform
X_scaled = scaler.fit_transform(X)

# Convert back to DataFrame for readability
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("Scaled feature sample:")
X_scaled.head()


# STEP 4 - Encoding categorical features

# In this dataset, categorical features are already numeric (SEX, EDUCATION, MARRIAGE, PAY_X)
# So no need for one-hot encoding. But we can cast them as categorical if needed.
categorical_features = ["SEX", "EDUCATION", "MARRIAGE"]

for col in categorical_features:
    X_scaled[col] = X_scaled[col].astype("category")

print("Categorical features encoded as categories (no dummy variables created).")


# SETP 5 - Split the dataset

from sklearn.model_selection import train_test_split

# Split data (70% training, 30% testing)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Training set size: {X_train.shape}")
print(f"Test set size: {X_test.shape}")


# STEP 6 - Final preprocessing summary

print("""
Preprocessing completed successfully:

- Categorical anomalies fixed (EDUCATION, MARRIAGE)
- ID column removed
- Data normalized with StandardScaler
- Train/Test split performed (70/30)
- Dataset ready for model training
""")

```

```python
# Question 3: Problem formalisation

# STEP 1 - Understanding the task

print("""
Problem formalisation:

The goal is to predict whether a client will default on their next credit card payment.

- Target variable: `default.payment.next.month`
    ➤ 1 = client will default
    ➤ 0 = client will not default

- Input features (X):
    ➤ Demographic data (SEX, AGE, EDUCATION, MARRIAGE)
    ➤ Credit information (LIMIT_BAL)
    ➤ Past payment history (PAY_0 to PAY_6)
    ➤ Billing and payment amounts (BILL_AMT1-6, PAY_AMT1-6)

This is a **supervised binary classification problem**.
""")


# STEP 2 - Variable types

categorical_features = ["SEX", "EDUCATION", "MARRIAGE"]
numerical_features = [col for col in X_scaled.columns if col not in categorical_features]

print("Categorical features:", categorical_features)
print("Numerical features:", numerical_features[:10], "...")


# STEP 3 - Target imbalance

import matplotlib.pyplot as plt
import seaborn as sns

# Calcul du pourcentage de chaque classe (0 = pas de défaut, 1 = défaut)
target_distribution = y.value_counts(normalize=True) * 100
print("Target distribution (%):\n", target_distribution)

# Visualisation de la répartition des classes
sns.countplot(x=y)
plt.title("Distribution of Target Variable (Default vs No Default)")
plt.xlabel("Default (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.show()

```

Observation:
There is a clear imbalance in the dataset:
- Around 78% of clients do NOT default
- Around 22% of clients DO default

This imbalance must be taken into account during model training (using class weights or resampling).


Since the dataset is imbalanced, accuracy alone is not sufficient.

We will use the following evaluation metrics:

- **Accuracy**: general performance of the model
- **Precision**: proportion of predicted defaults that are correct
- **Recall (Sensitivity)**: ability to detect true defaults
- **F1-score**: balance between precision and recall
- **ROC-AUC**: overall discriminatory power of the model

These metrics are especially useful for **imbalanced binary classification**.


Problem Summary:

- Type of problem: Supervised learning (Binary Classification)
- Target variable: default.payment.next.month
- Input: 23 explanatory variables (credit, demographic, and payment history)
- Evaluation metrics: Accuracy, Precision, Recall, F1-score, ROC-AUC
- Specific challenge: Class imbalance (78% vs 22%)

```python
# Question 4: Baseline model

# STEP 1 - Model selection rationale

print("""
Baseline Model Choice: Logistic Regression

Justification:
- Simple, interpretable and fast to train
- Well-suited for binary classification problems
- Provides probability outputs for ROC-AUC evaluation
- Serves as a reference for more advanced models (e.g., Random Forest, XGBoost)

We'll train it using the preprocessed and scaled dataset (X_train, y_train).
""")


# -STEP 2 - Train logistic regression baseline model

from sklearn.linear_model import LogisticRegression

# Initialize model (with class weights to handle imbalance)
log_reg = LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)

# Train model
log_reg.fit(X_train, y_train)

# Predictions
y_pred = log_reg.predict(X_test)
y_proba = log_reg.predict_proba(X_test)[:, 1]

print("Logistic Regression model trained successfully.")


# STEP 3 - Evaluate baseline model

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, ConfusionMatrixDisplay

# Compute metrics
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_proba)

print(f"""
Model Performance (Baseline - Logistic Regression):
-----------------------------------------------------
Accuracy   : {acc:.3f}
Precision  : {prec:.3f}
Recall     : {rec:.3f}
F1-score   : {f1:.3f}
ROC-AUC    : {roc_auc:.3f}
""")

# Confusion matrix
ConfusionMatrixDisplay.from_estimator(log_reg, X_test, y_test, cmap='Blues')
plt.title("Confusion Matrix – Logistic Regression (Baseline)")
plt.show()


# STEP 4 - ROC curve

from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_estimator(log_reg, X_test, y_test)
plt.title("ROC Curve – Logistic Regression (Baseline)")
plt.show()
```

```python
# Improved Decision Tree with RandomizedSearchCV

from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix
)
from scipy.stats import randint
import seaborn as sns
import matplotlib.pyplot as plt

X = df.drop(columns=["ID", "default.payment.next.month"])
y = df["default.payment.next.month"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# Basic model
dt = DecisionTreeClassifier(random_state=42)

#  Hyperparameter search space
param_distributions = {
    "criterion": ["gini", "entropy"],
    # we use ranges rather than fixed values
    "max_depth": [3, 4, 5, 6, 8, None],
    "min_samples_split": randint(2, 101),    # 2 to 100
    "min_samples_leaf": randint(1, 101),     # 1 to 100
    "class_weight": [None, "balanced"]
}

# RandomizedSearchCV with cross-validation
random_search = RandomizedSearchCV(
    estimator=dt,
    param_distributions=param_distributions,
    n_iter=50,            # number of randomly drawn combinations
    scoring="f1",
    cv=5,
    n_jobs=-1,
    random_state=42,
    verbose=1
)

random_search.fit(X_train, y_train)

print("Meilleurs hyperparamètres trouvés :")
print(random_search.best_params_)
print(f"Meilleur score CV (métrique f1) : {random_search.best_score_:.3f}")

# We get the best model
best_dt = random_search.best_estimator_

# Evaluation on the test set
y_pred = best_dt.predict(X_test)
y_proba = best_dt.predict_proba(X_test)[:, 1]

accuracy  = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall    = recall_score(y_test, y_pred)
f1        = f1_score(y_test, y_pred)
roc_auc   = roc_auc_score(y_test, y_proba)

print("\n=== Decision Tree amélioré (RandomizedSearchCV) – performances sur le test set ===")
print(f"Accuracy  : {accuracy:.3f}")
print(f"Precision : {precision:.3f}")
print(f"Recall    : {recall:.3f}")
print(f"F1-score  : {f1:.3f}")
print(f"ROC-AUC   : {roc_auc:.3f}")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4, 3))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Prédit")
plt.ylabel("Vrai")
plt.title("Decision Tree amélioré – Matrice de confusion")
plt.show()

# Tree visualization (truncated for readability)
plt.figure(figsize=(20, 8))
plot_tree(
    best_dt,
    feature_names=X.columns,
    class_names=["no default", "default"],
    filled=True,
    rounded=True,
    max_depth=3   # only the first 3 levels are displayed
)
plt.title("Decision Tree amélioré – Structure (niveaux 0 à 3)")
plt.show()
```

=========== Advanced KNN ===========



```python
# KNN optimized (RandomizedSearch)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix
)
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Define X / y + split
X = df.drop(columns=["ID", "default.payment.next.month"])
y = df["default.payment.next.month"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# Base KNN model
# We optimize this model using RandomizedSearchCV.
knn = KNeighborsClassifier()

# Hyperparameter search space
# n_neighbors : number of neighbors considered (K)
# weights     : "uniform" = all neighbors have equal weight,
#               "distance" = closer neighbors have more influence
# p           : distance metric (1=Manhattan, 2=Euclidean)
param_dist = {
    "n_neighbors": np.arange(3, 30),
    "weights": ["uniform", "distance"],
    "p": [1, 2]
}

# RandomizedSearchCV
# Randomized search tries a limited number of random combinations,
# much faster than GridSearch (which tries every single combination).
random_search = RandomizedSearchCV(
    estimator=knn,
    param_distributions=param_dist,
    n_iter=20,          # number of random combinations to test
    scoring="f1",       # F1-score = good metric for imbalanced data
    cv=3,               # 3-fold cross-validation (faster than cv=5)
    n_jobs=-1,          # use all CPU cores
    verbose=1,
    random_state=42
)

random_search.fit(X_train, y_train)

print("Best hyperparameters found:")
print(random_search.best_params_)
print(f"Best CV score (F1) : {random_search.best_score_:.3f}")

# Retrieve the best model found during search
best_knn = random_search.best_estimator_

# Evaluation on the test set
y_pred = best_knn.predict(X_test)

try:
    y_proba = best_knn.predict_proba(X_test)[:, 1]
except:
    y_proba = None

accuracy  = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall    = recall_score(y_test, y_pred)
f1        = f1_score(y_test, y_pred)

print("\n=== KNN optimized – Test set performance ===")
print(f"Accuracy  : {accuracy:.3f}")
print(f"Precision : {precision:.3f}")
print(f"Recall    : {recall:.3f}")
print(f"F1-score  : {f1:.3f}")

if y_proba is not None:
    roc_auc = roc_auc_score(y_test, y_proba)
    print(f"ROC-AUC   : {roc_auc:.3f}")

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(4,3))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True label")
plt.title("KNN optimized – Confusion matrix")
plt.show()


```

=========== Algorithm Outside Of The Course : MLP ===========

```python
# MLP (Neural Network)

import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, confusion_matrix
)

import seaborn as sns
import matplotlib.pyplot as plt

# Define X / y + train-test split
X = df.drop(columns=["ID", "default.payment.next.month"])
y = df["default.payment.next.month"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,
    random_state=42,
    stratify=y
)

# Convert to numpy arrays (Keras can also work with pandas, but numpy is explicit)
X_train_np = X_train.values
X_test_np  = X_test.values
y_train_np = y_train.values
y_test_np  = y_test.values

# Compute class weights (balanced) from training labels only
# This increases the penalty for mistakes on the minority class (default = 1)
classes = np.unique(y_train_np)
weights = compute_class_weight(class_weight="balanced", classes=classes, y=y_train_np)
class_weight = {int(c): float(w) for c, w in zip(classes, weights)}
print("Class weights used:", class_weight)

# Define the MLP architecture
input_dim = X_train_np.shape[1]

model = keras.Sequential([
    layers.Input(shape=(input_dim,)),
    layers.Dense(64, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(32, activation="relu"),
    layers.Dropout(0.3),
    layers.Dense(1, activation="sigmoid")
])

# Compile the model
# AUC is more informative than accuracy on imbalanced datasets
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss="binary_crossentropy",
    metrics=[tf.keras.metrics.AUC(name="auc")]
)

# Train the model with EarlyStopping on validation AUC
early_stop = keras.callbacks.EarlyStopping(
    monitor="val_auc",
    mode="max",
    patience=5,
    restore_best_weights=True
)

history = model.fit(
    X_train_np, y_train_np,
    validation_split=0.2,
    epochs=50,
    batch_size=256,
    callbacks=[early_stop],
    class_weight=class_weight,
    verbose=1
)

# Evaluation on the test set
y_proba = model.predict(X_test_np).ravel()
y_pred = (y_proba >= 0.5).astype(int)

accuracy  = accuracy_score(y_test_np, y_pred)
precision = precision_score(y_test_np, y_pred, zero_division=0)
recall    = recall_score(y_test_np, y_pred, zero_division=0)
f1        = f1_score(y_test_np, y_pred, zero_division=0)
roc_auc   = roc_auc_score(y_test_np, y_proba)

print("\n=== MLP (class_weight) – Test set performance ===")
print(f"Accuracy  : {accuracy:.3f}")
print(f"Precision : {precision:.3f}")
print(f"Recall    : {recall:.3f}")
print(f"F1-score  : {f1:.3f}")
print(f"ROC-AUC   : {roc_auc:.3f}")

# Confusion matrix
cm = confusion_matrix(y_test_np, y_pred)
plt.figure(figsize=(4, 3))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("True label")
plt.title("MLP (class_weight) – Confusion matrix")
plt.tight_layout()
plt.show()

# Plot training history (AUC)
plt.figure(figsize=(5, 3))
plt.plot(history.history["auc"], label="Train AUC")
plt.plot(history.history["val_auc"], label="Validation AUC")
plt.xlabel("Epoch")
plt.ylabel("AUC")
plt.title("MLP (class_weight) – Training and validation AUC")
plt.legend()
plt.tight_layout()
plt.show()

```

```python
# Voting Classifier

from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score
)

import pandas as pd

if "df_results" not in globals():
    df_results = pd.DataFrame(
        columns=["Modèle", "Accuracy", "Precision", "Recall", "F1-score", "AUC"]
    )

# Define base models (light versions to avoid long training)
lr = LogisticRegression(
    max_iter=1000,
    class_weight="balanced",
    random_state=42
)

rf = RandomForestClassifier(
    n_estimators=300,
    class_weight="balanced",
    random_state=42,
    n_jobs=-1
)

gb = GradientBoostingClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=3,
    random_state=42
)

xgb = XGBClassifier(
    objective="binary:logistic",
    eval_metric="logloss",
    n_estimators=400,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    n_jobs=-1
)

voting_clf = VotingClassifier(
    estimators=[
        ("lr", lr),
        ("rf", rf),
        ("gb", gb),
        ("xgb", xgb)
    ],
    voting="soft",
    weights=[1, 2, 2, 2],  # give more importance to strong models
    n_jobs=-1
)

# Train
voting_clf.fit(X_train, y_train)

# Evaluate
y_pred = voting_clf.predict(X_test)
y_proba = voting_clf.predict_proba(X_test)[:, 1]

accuracy  = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, zero_division=0)
recall    = recall_score(y_test, y_pred, zero_division=0)
f1        = f1_score(y_test, y_pred, zero_division=0)
roc_auc   = roc_auc_score(y_test, y_proba)

print("=== Voting Classifier (Soft) ===")
print(f"Accuracy  : {accuracy:.3f}")
print(f"Precision : {precision:.3f}")
print(f"Recall    : {recall:.3f}")
print(f"F1-score  : {f1:.3f}")
print(f"ROC-AUC   : {roc_auc:.3f}")

# Add results directly to existing table
df_results.loc[len(df_results)] = [
    "Voting Classifier (Soft)",
    accuracy,
    precision,
    recall,
    f1,
    roc_auc
]

df_results

```
